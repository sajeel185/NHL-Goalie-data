```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("goalie_data.csv")
print(df)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['shots'], color='purple')
plt.ylabel('Values')
plt.title('Shots')

plt.subplot(1, 2, 2)
plt.plot(df['saves'], color='green')
plt.ylabel('Values')
plt.title('Saves')


plt.show()
```

These are two line plots showing that the number of saves a goalie in the NHL makes, is dependent upon the number of shots taken against him.  This confirms that there aren't any irregularities in our data points.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(df['timeOnIce'])
plt.title("Time on Ice")
plt.xlabel("Minutes")
plt.ylabel("Value")

plt.subplot(1, 2, 2)
plt.scatter(x=df['timeOnIce'], y=df['saves'])
plt.xlabel("Time on Ice (Mins)")
plt.ylabel("Saves")
plt.title('Time on Ice vs. Saves')

plt.show()
```

The first plot is a histgram of the "minutes on ice" for each goalie.  It shows that the majority of goalies in the NHL played between 3500 to 4200 minutes in a season.  From this we can infer that in general that the data set we have is fair in its assessment, and that there is no bias towards some goalies data points because of an unfair advantage in playing time on ice.  The scatter plot on the right positively correlates the amount of playing time on ice to the number of saves a goalie has over the course of a season.


```python
y = df["evenSaves"]
z = df["shortHandedSaves"]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(y)
plt.ylabel('Values')
plt.title('evenSaves')

plt.subplot(1, 2, 2)
plt.boxplot(z)
plt.ylabel('Values')
plt.title('shortHandedSaves')


plt.show()
```

These are a set of two box plot.  The first box plot titled "evenSaves" shows that the number of even saves that goalies had in the NHL and the second box plot titled "shortHandedSaves" shows the number of short handed saves that goalies had in the NHL.  By looking at the first box plot, we see that the number of even saves is in a dramatically higher range than the number of short handed saves.  From this we could suggest that players shoot towards a goalie's short hand to increase their chances of scoring.
